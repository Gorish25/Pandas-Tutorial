## Replace

- replacing all 5 with 21

```python
# replacing all 5 with 21
csv1.replace(to_replace='5',value=21)
```

```python
csv1 = pd.read_csv('test1.csv')
csv1
```

- replacing any value using replace

```python
# replacing any string with other string
csv1.replace(to_replace='Financial ratios', value='Python')
```

```python
# replacing any number of data with other number
csv1.replace(['H01','H04','H05','H07','H08'],'Java')
```

```python
# replacing all data with python using regex
csv1.replace('[A-Za-z0-9]','Python',regex=True)
```

```python
# replacing all specific column data using dictionary
csv1.replace({'Variable_name':'[A-Za-z]'},"Hello Moto",regex=True)
```

```python
# replacing specific data using replace(method='ffill')
csv1.replace('ZZ11',method='ffill')
```

```python
# replacing specific data using replace(value,method='bfill')
csv1.replace(46,method='bfill')
```

```python
# replacing specific data using replace(value,method,limit)
csv1.replace('40',method='ffill',limit=2)
```

```python
# inplace = it converts the original data and do not form a copy
csv1.replace('99999','python',inplace=True)
csv1
```
