```python
import pandas as pd
```

- for reshaping a pandas file 2 functions are used 
- pivot and melt

## Melt()

```python
var1 = pd.DataFrame({'Days':[1,2,3,4,5,6],
                    'Eng':[10,12,11,14,15,19],
                    'Math':[17,18,12,15,19,20]
                    })
var1
```

```python
# reshaping pandas file vertically
var1.melt()
```

```python
# converting any id to column
pd.melt(var1, id_vars=['Eng'])
```

```python
# arranging correctly
pd.melt(var1, id_vars=['Days'])
```

```python
# changing variable name
pd.melt(var1, id_vars=['Days'], var_name='Python')
```

## pivot()

```python
var1 = pd.DataFrame({'Days':[1,2,3,4,5,6],
                     'st_name':['a','b','c','d','e','f'],
                    'Eng':[10,12,11,14,15,19],
                    'Math':[17,18,12,15,19,20]
                    })
var1
```

```python
var1.pivot(index="Days",columns='st_name')
```

```python
# accessing a specific subject data
var1.pivot(index='Days',columns='st_name',values='Eng')
```

```python
# pivot_table => finding mean
var1 = pd.DataFrame({'days':[1,1,1,1,2,2],
                    'st_name':['a','b','a','a','b','b'],
                     'eng':[10,15,12,14,11,13],
                     'maths':[11,14,12,16,17,14]
                    })

var1
```

```python
# finding mean
var1.pivot_table(index='st_name',columns='days',aggfunc='mean')
```

```python
# finding sum
var1.pivot_table(index='st_name', columns='days', aggfunc='sum')
```

```python
# finding all means
var1.pivot_table(index='st_name', columns='days', aggfunc='sum', margins='True')
```
