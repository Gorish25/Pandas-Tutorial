```python
import pandas as pd
```

## dropna & fillna

```python
csv1 = pd.read_csv('test1.csv')
csv1
```

- dropping missing data row

```python
csv1.dropna()
```

- dropping missing column data

```python
csv1.dropna(axis=1)
```

- removing all blank field using dropna(how='any')

```python
csv2 = pd.read_csv('test2.csv')
csv2.dropna(how='any')
```

- removing a special row with no value and filling all other rows missing only one value with NaN

```python
csv2.dropna(how='all')
```

- removing any particular field NaN value using dropna subset parameter

```python
csv2.dropna(subset=['Year'])
```

- removing all NaN values and creating a new database using dropna inplace attribute

```python
csv2.dropna(inplace=True)
csv2
```

- thresh attribute of dropna => it is used when we want to remove a row having (thresh=n) NaN value (not upper than n)

```python
csv3 = pd.read_csv('test2.csv')
csv3.dropna(thresh=2)
```

- filling NaN values

```python
csv3.fillna('Python')
```

- filling a particular data in a particular column

```python
csv3.fillna({'Year':'Java','Value':500000})
```

- if we want to fill NaN with its forward occuring data then we use fillna(method='ffill')
- if we want to fill NaN with its fillna(method='bfill')

```python
csv3.fillna(method='ffill')
```

```python
csv3.fillna(method='bfill')
```

```python
# filling using axis
csv3.fillna(method='bfill',axis=1)
```

```python
csv3.fillna(method='ffill',axis=1)
```

- using inplace attribute of fillna

```python
csv3.fillna(12,inplace=True)
csv3
```

- filling data upto a given limit of rows

```python
csv4 = pd.read_csv('test2.csv')
csv4.fillna('Python',limit=2)
```
